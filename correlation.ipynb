```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
from matplotlib.pyplot import figure
from functools import reduce
```

```python
data = pd.read_csv('seeker_satisfaction_1000_thread_containing_content_update2.csv')
```

```python
corr = data[['satisfaction', 'post_num_comments', 'post_score', 'post_upvote_ratio', 'post_tenure', 'familiarity',
               'post_IS', 'post_ES', 'comment_score', 'comment_IS', 'comment_ES', 'reply_length', 'post_between_comment',
               'comment_between_reply', 'IS_difference', 'ES_difference', 'difference_sum']].corr(method='pearson')
plt.figure(figsize=(20,20))
ax = sns.heatmap(corr, annot=True,
            linewidths=.5,
            cbar_kws={"shrink": .5},
            vmin = -1,vmax = 1,
            square=True, cmap='RdYlGn_r')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```
